```python
import pandas as pd
import numpy as np

# 1. Create a dummy demographic dataset representing employment data
data = {
    'Employee_ID': range(1, 11),
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'Years_Experience': [2, 3, 5, 6, 1, 7, 4, 8, 2, 5],
    'Monthly_Salary_PKR': [75000, 85000, 110000, 140000, 60000, 155000, 95000, 170000, 78000, 125000]
}

df = pd.DataFrame(data)
print("--- Raw Demographic Data ---")
print(df)
print("\n")

# 2. Mathematical Analysis: Calculate average salary by Gender
avg_salary = df.groupby('Gender')['Monthly_Salary_PKR'].mean()

print("--- Statistical Metrics: Average Salary by Gender ---")
print(avg_salary)
print("\n")

# 3. Calculate Pay-Gap Ratio (A basic mathematical metric for equity analysis)
female_avg = avg_salary['Female']
male_avg = avg_salary['Male']
pay_gap_ratio = (female_avg / male_avg) * 100

print(f"Mathematical Insight: Females earn {pay_gap_ratio:.2f}% of what males earn for similar experience spreads.")
```
